Когда мы разбирали PMI мы использовали такую функцию:

```python
import re
corpus= corpus.lower()
corpus = [w for w in corpus.split() if re.match(r'[\w\d]+', w)]

```

```python
def score_bigrams(corpus, scorer):
    unigrams = Counter(corpus)
    bigrams = Counter(zip(corpus, corpus[1:]))
    total = len(corpus)
    
    scores = {}
    for (w1, w2), bigram_count in bigrams.items():
        scores[(w1, w2)] = scorer(
            unigrams[w1], 
            unigrams[w2], 
            bigram_count, 
            total
        )
    return scores
```

```python
def scorer_simple(word_count_a, word_count_b, bigram_count, *args):
    try:
        score = bigram_count/((word_count_a+word_count_b))
    except ZeroDivisionError:
        return 0
    return score
```

```python
def scorer_pmi(word_count_a, word_count_b, bigram_count, total_tokens):
    import math
    if bigram_count == 0:
        return float("-inf")
    p_x = word_count_a / total_tokens
    p_y = word_count_b / total_tokens
    p_xy = bigram_count / total_tokens
    return math.log2(p_xy / (p_x * p_y))
```

```python
scores_simple = score_bigrams(corpus, scorer_simple)
scores_pmi = score_bigrams(corpus, scorer_pmi)
```

```python
top_simple = sorted(scores_simple.items(), key=lambda x: -x[1])[:10]
top_pmi = sorted(scores_pmi.items(), key=lambda x: -x[1])[:10]
```

```python

print("Top-10 по simple:")
for pair, score in top_simple:
    print(pair, score)

print("\nTop-10 по PMI:")
for pair, score in top_pmi:
    print(pair, score)
```

```python
import math

def scorer_pmi_log(word_count_a, word_count_b, bigram_count, total_tokens):
    if bigram_count == 0:
        return float("-inf")
    log_p_x = math.log2(word_count_a) - math.log2(total_tokens)
    log_p_y = math.log2(word_count_b) - math.log2(total_tokens)
    log_p_xy = math.log2(bigram_count) - math.log2(total_tokens)
    return log_p_xy - (log_p_x + log_p_y)
```
